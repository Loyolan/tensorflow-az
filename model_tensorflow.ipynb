```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import pandas as pd
import sys
```

### DATASETS

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
```

```python
(images, targets), (_, _) = fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split
images_train, images_test, targets_train, targets_test = train_test_split(images, targets, test_size=0.2, random_state=1)
print(images_train.shape, targets_train.shape)
print(images_test.shape, targets_test.shape)
```

```python
plt.figure(figsize=(14, 4))
for i in range(5):
    plt.subplot(1, 20, i+1)
    plt.imshow(images_train[i])
plt.show()
```

```python
targets_name = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
plt.figure(figsize=(14, 4))
for i in range(10):
    plt.subplot(2, 10, i+1)
    plt.title(targets_name[targets[i]])
    plt.imshow(images[i])
plt.show()
```

### CREATE A MODEL

```python
model = tf.keras.models.Sequential()
# applatir l'image
model.add(tf.keras.layers.Flatten(input_shape=[28, 28]))
images[0:1].shape
```

```python
model = tf.keras.models.Sequential()
# APPLATIR L'IMAGE
model.add(tf.keras.layers.Flatten(input_shape=[28, 28]))
# 1ERE COUCHE
model.add(tf.keras.layers.Dense(256, activation='relu'))
# 2EME COUCHE
model.add(tf.keras.layers.Dense(128, activation='relu'))
# SORTIE
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

### RESUME DU MODEL

```python
model.summary()
```

### COMPILE MODEL

```python
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

### TRAIN MODEL

```python
history = model.fit(images_train, targets_train, epochs=20, validation_split=0.2)
```

```python
loss_curve = history.history['loss']
acc_curve = history.history['accuracy']
loss_val_curve = history.history['val_loss']
acc_val_curve = history.history['val_accuracy']
plt.figure(figsize=(14, 4))
plt.subplot(1, 2, 1)
plt.plot(loss_curve)
plt.plot(loss_val_curve)
plt.title("LOSS")
plt.subplot(1, 2, 2)
plt.plot(acc_curve)
plt.plot(acc_val_curve)
plt.title("ACCURACY")
plt.show()
```
